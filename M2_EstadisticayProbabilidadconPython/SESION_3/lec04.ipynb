## 1. Vectores y Escalares

>Un __vector de dimensión $n$ o vector en $\mathbb{R}^n$, es un conjunto ordenado de $n$ elementos__.
<br><br>
$$
\mathbb{R}^n =  \left\lbrace  \ 
\begin{pmatrix}
\xi_1\\  \xi_2\\ \vdots \\ \xi_n
\end{pmatrix}
: \xi_1 ,\, \xi_2,\, ...\, , \xi_n \   \in \ \mathbb{R}  \right\rbrace
$$
<br>
Y cada elemento o __componente__ ($\xi_1 ,\, \xi_2,\, ...\, , \xi_n$), puede ser un número real, que en su *conjunto ordenado* se puede representar como un __vector columna de dimensión $n$ ($v ∈ \mathbb{R}^n$)__, así:
<br><br>
$$v = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix}$$
<br>
Donde, $v_1, v_2, \ldots, v_n$ son los __componentes__ del vector de dimensión u orden $n$, y el número de elementos o componentes en el vector representa la __dimensión u orden del vector__, 
<br>


Ejemplo: En $\mathbb{R}^2$,  el vector $$u = \begin{bmatrix} 4 \\ 2 \end{bmatrix}$$

Por ejemplo, la forma de un vector en $\mathbb{R}^3$ ($v ∈ \mathbb{R}^3$), es decir un vector situado en el espacio tridimensional de los números reales o en coordenadas reales, sería así:

$$v = \begin{bmatrix} 2 \\ 4 \\ 5 \end{bmatrix}$$

Este vector tiene dimension $n = 3$ porque tiene tres componentes, y cada uno de estos componentes puede corresponder a una __coordenada en el espacio tridimensional $x,y,z$__; donde: $x=2$, $y=4$, y $z=5$.
<br>
Este vector $v(2, 4, 5)$ es un punto en el espacio tridimensional o espacio vectorial ($v ∈ \mathbb{R}^3$), y se representa como un segmento de recta orientada del origen a la coordenada correspondiente ($x=2$, $y=4$, $z=5$); también representado como $\vec v\; = \;\left( {2,\;4,\;5} \right)$:

La **norma** (euclidiana) de un vector $$v = (v_1, v_2, \ldots, v_n)^t$$ en un espacio de dimensión $n$ se define como la raíz cuadrada de la suma de los cuadrados de sus componentes. Matemáticamente, se expresa como:

$$
||v|| = \sqrt{v_1^2 + v_2^2 + \ldots + v_n^2}
$$

Esta es una generalización de la fórmula de la distancia euclidiana en un espacio de más de dos o tres dimensiones.
