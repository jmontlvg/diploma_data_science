### `random.choices()`: para muestra *k* valores de una lista con reemplazo

```python
apellidos=["Saenz","León","Mosqueda","Dante","Curiel","Abad","Miranda","Carrasco","Gutiérrez","Cid","Hernández"]
```

```python
seed(1)
sampled_values=random.choices(apellidos,k=len(empleados))#me regresa una lista
sampled_values
```

```python
df_apellidos=pd.DataFrame({'apellido':sampled_values})
df_apellidos
```

```python
df_apellidos['id']=range(1,len(df_apellidos)+1)
df_apellidos
```
