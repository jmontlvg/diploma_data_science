>Un __vector de dimensión $n$ o vector en $\mathbb{R}^n$, es un conjunto ordenado de $n$ elementos__.
<br><br>
$$
\mathbb{R}^n =  \left\lbrace  \ 
\begin{pmatrix}
\xi_1\\  \xi_2\\ \vdots \\ \xi_n
\end{pmatrix}
: \xi_1 ,\, \xi_2,\, ...\, , \xi_n \   \in \ \mathbb{R}  \right\rbrace
$$
<br>
Y cada elemento o __componente__ ($\xi_1 ,\, \xi_2,\, ...\, , \xi_n$), puede ser un número real, que en su *conjunto ordenado* se puede representar como un __vector columna de dimensión $n$ ($v ∈ \mathbb{R}^n$)__, así:
<br><br>
$$v = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix}$$
<br>
Donde, $v_1, v_2, \ldots, v_n$ son los __componentes__ del vector de dimensión u orden $n$, y el número de elementos o componentes en el vector representa la __dimensión u orden del vector__, 
<br>


Ejemplo: En $\mathbb{R}^2$,  el vector $$u = \begin{bmatrix} 4 \\ 2 \end{bmatrix}$$

* ## Suma de vectores:
    
    Es una operación que toma dos vectores de ${\mathbb{R}^n}$ y devuelve un vector de ${\mathbb{R}^n}$: 
    $$
    \vec v + \vec w = \left( {{v_x} + {w_x}\;,\;\;{v_y} + {w_y}\;\;,\;\;{v_z} + {w_z}} \right)^t
    $$
    <br>
    Se pueden sumar vectores del mismo orden o dimensión (misma cantidad de componentes/elementos), y se suman componente por componente:
    $$
    \vec v + \vec w = \begin{bmatrix} {v_x} \\ {v_y} \\ {v_z} \end{bmatrix} + \begin{bmatrix} {w_x} \\ {w_y} \\ {w_z} \end{bmatrix} = \begin{bmatrix} {v_x} + {w_x} \\ {v_y} + {w_y} \\ {v_z} + {w_z} \end{bmatrix}
    $$
    <br>
    Sustituyendo los valores de los vectores:
    $$
    \vec v + \vec w = \begin{bmatrix} 2 \\ 4 \\ 5 \end{bmatrix} + \begin{bmatrix} 1 \\ 3 \\ 7 \end{bmatrix} = \begin{bmatrix} 3 \\ 7 \\ 12 \end{bmatrix}
    $$
    <br>
    El resultado de la suma de dos vectores, nos da otro vector determinada por sus elementos:
    
    * Módulo o norma: valor númerico del vector, es su longitud.
    * Dirección: la línea recta sobre la que está situado.
    * Sentido: la orientación de dicho vector, indicado por su flecha.
    * Punto de aplicación: el origen de dicho vector.
    
    <br>
    Propiedades principales de la suma de vectores:
    
    * Es **conmutativa**, lo que significa que ${v} + {w} = {w} + {v}$
    * Es **asociativa**, lo que significa que ${u} + ({v} + {w}) = ({u} + {v}) + {w}$
    
    <br>

* ## Producto Punto:
    También llamado producto escalar de dos vectores $v$ y $w$ en ${\mathbb{R}^n}$.
    
    Se obtiene sumando el producto de los componentes correspondientes de cada vector, donde dicho producto será un escalar ($k$):
    $$
    v \cdot  w = \sum_i{v_i \cdot w_i} = {{v_1} ⋅ {w_1}\; + \;\;{v_2} ⋅ {w_2}\;\; + \;\, ...\,\; + \;\;{v_n} ⋅ {w_n}} = {k}
    $$
    
    También se puede anotar el producto de las siguientes maneras:
    * $\vec v \cdot \vec w$ 
    * $(v|w)$
    * $⟨v|w⟩$
    
    <br>
    El resultado del producto punto de dos vectores, nos da un escalar o una magnitud escalar determinada por un número y una unidad:
    
    $$
    \vec v \cdot \vec w = \begin{bmatrix} {v_x} \\ {v_y} \\ {v_z} \end{bmatrix} \cdot \begin{bmatrix} {w_x} \\ {w_y} \\ {w_z} \end{bmatrix} = (v_x \times {w_x}) + (v_y \times w_y) + (v_z \times w_z)
    $$
    <br>
    Sustituyendo los valores de los vectores:
    $$
    \vec v \cdot \vec w = \begin{bmatrix} 2 \\ 4 \\ 5 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 3 \\ 7 \end{bmatrix} = (2 \times 1) + (4 \times 3) + (5 \times 7) = 49
    $$
       
    <br>
    Propiedades principales del producto punto:
    
    * Es **conmutativo**: ${v} \cdot {w} = {w} \cdot {v}$
    * **NO es asociativo** entre vectores: $({u} \cdot {v}) \cdot {w} ≠ {u} \cdot ({v} \cdot {w})$
    * Pero, **sí es asociativo con respecto a la multiplicación con un escalar**: ${k} \times ({u} \cdot {v}) = ({k} \times {u}) \cdot {v} = {u} \cdot ({k} \times {v})$
    * Es **distributivo** sobre la suma de vectores: ${u} \cdot ({v} + {w}) = {u} \cdot {v} + {u} \cdot {w}$
    
    <br>

# Concepto relacionado: Correlación

La correlación entre dos vectores   $ \mathbf{a} = (a_1, a_2, \ldots, a_n)$  y $ \mathbf{b} = (b_1, b_2, \ldots, b_n) $, se puede calcular usando la fórmula de  correlación de Pearson: 
$$
\rho(\mathbf{a}, \mathbf{b}) = \frac{\sum_{i=1}^{n} (a_i - \bar{a})(b_i - \bar{b})}{\sqrt{\sum_{i=1}^{n} (a_i - \bar{a})^2} \sqrt{\sum_{i=1}^{n} (b_i - \bar{b})^2}}
$$

Donde:
- $ \bar{a} $ es el promedio de los elementos de $ \mathbf{a} $. 

- $ \bar{b} $ es el promedio de los elementos de $ \mathbf{b} $.

Es decir: $\bar{a} = \frac{1}{n} \sum_{i=1}^{n} a_i$  y  $\bar{b} = \frac{1}{n} \sum_{i=1}^{n} b_i$

Esta fórmula mide la relación lineal entre los dos vectores. Un valor de correlación cercano a 1 indica una fuerte correlación positiva, cercano a -1 indica una fuerte correlación negativa, y cercano a 0 indica que no hay una correlación lineal fuerte.

En términos del producto punto:
$$
\rho(\mathbf{a}, \mathbf{b}) = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|}
$$

donde se asume que a los vectores ya les restamos su media (vectores distintos de cero)

## Vectores Ortogonales

```python
e1_norm=la.norm(e1)
e2_norm=la.norm(e2)

cos=e1e2/(e1_norm*e2_norm)
rad=np.arccos(cos) #para sacar el ángulo en radianes
grados=np.rad2deg(rad)
print(grados)
print(rad)
```

* Dos vectores son **ortogonales** si su **producto punto es cero**
* Cuando esto pasa, el ángulo entre ello es de 90 grados, que equivale a $\pi/2$ radianes.
* Si dos vectores son ortogonales nos aportan información 'independiente" o muy distinta. Es decir: no hay redundancia en la información que nos dan.
* Esta propiedad es muchas veces veces ideal en el análisis de datos.

# Combinación Lineal

Una combinación lineal de los vectores  $\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_n$  se define como una expresión de la forma \
$$
c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \cdots + c_n\mathbf{v}_n \\
$$

donde $c_1, c_2, \ldots, c_n$  son escalares



#Ejemplo:

$$
    2v + w = 2\begin{bmatrix} 2 \\ 4 \\ 5 \end{bmatrix} + \begin{bmatrix} 1 \\ 3 \\ 7 \end{bmatrix} = \begin{bmatrix} 5 \\ 11 \\ 17 \end{bmatrix}
    $$
    <br>

### Independencia Lineal

#### Definición 
Matemáticamente, un conjunto de vectores $ \mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_n $ en un espacio vectorial es **linealmente independiente** si la única solución a la ecuación

$$
\ c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \cdots + c_n\mathbf{v}_n = \mathbf{0} \
$$

es que todos los coeficientes $c_1, c_2, \ldots, c_n$ sean cero. Esto significa que ningún vector en el conjunto puede ser expresado como una combinación lineal de los otros.

La independencia lineal entre vectores es un concepto fundamental en álgebra lineal. Proporciona una manera de entender cómo los vectores se relacionan entre sí en términos de su capacidad para construir o no un espacio vectorial.

#### Intuición
La idea detrás de la independencia lineal es determinar si un conjunto de vectores puede generar un espacio vectorial sin "sobreponer" sus contribuciones. En otras palabras, se trata de ver si cada vector en el conjunto aporta una "dirección" única que no puede ser replicada o construida por una combinación de los otros vectores en el conjunto.

Si imaginamos los vectores como flechas en un espacio, un conjunto de vectores es linealmente independiente si ninguna de estas flechas puede alcanzarse "caminando" a lo largo de las otras flechas usando una combinación de pasos hacia adelante y hacia atrás a lo largo de ellas.



### Dependencia lineal:
- Si la propiedad de arriba no se cumple, es decir existe al menos algún $c_i$ distinto de cero tal que la combinación lineal sea cero, entonces decimos que el conjunto de vectores es **linealmente dependiente**
- Esto implica que hay al menos un vector que puede expresarse como combinación lineal del resto
