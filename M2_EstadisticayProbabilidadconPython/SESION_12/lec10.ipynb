### Variable Aleatoria

Una variable aleatoria (v.a.) es esencialmente una **función que asigna números a los resultados de un experimento aleatorio.**

##### Ejemplo: 

1. Supongamos que lanzamos un dado. Podemos definir una v.a. $X$ para representar el resultado del lanzamiento. Aquí, $X$ es una v.a. discreta, ya que sólo puede tomar los valores enteros 1, 2, 3, 4, 5 o 6, correspondientes a los posibles resultados.
<br><br>
2. Supongamos estamos en una parada de autobus. Una v.a. en este caso podría ser:
$$
T = \text{tiempo de espera hasta que llegue el autobús.}
$$

En este caso, podríamos esperar 5 minutos, 7.25 minutos, 10 minutos y 34 segundos, etc. $T$ puede tomar cualquier valor en un continuo de tiempos y por tanto es una v.a. contínua.
<br><br>
Otros ejemplos:
<br><br>

3. $Y$ = Tiempo de espera en ser atendido en un mostrador

4. $Z$ = Número de veces que cae 'aguila' al lanzar una moneda.

5. $T$ = Tiempo de recuperación de un paciente tras recibir tratamiento.

Ejemplos de variables aleatorias discretas: 

1. **Lanzar una Moneda:**
   - **Experimento:** Lanzar una moneda justa.
   - **Variable Aleatoria Discreta:** El resultado del lanzamiento de la moneda.
   - **Explicación:** Definamos una variable $X$ tal que $X = 1$ si la moneda cae en 'aguila' y $X = 0$ si cae en 'sol'. Esta variable es discreta porque solo puede tomar dos valores distintos, 1 o 0, correspondientes a los dos posibles resultados del lanzamiento de la moneda.


2. **Número de Artículos Defectuosos:**
   - **Experimento:** Inspeccionar un lote de artículos en busca de defectos.
   - **Variable Aleatoria Discreta:** El número de artículos defectuosos en el lote.
   - **Explicación:** Supongamos  tenemos un lote de 100 artículos e inspeccionamos cada uno en busca de defectos. Sea $Z$  el recuento total de artículos defectuosos. $Z$ puede ser cualquier número entero entre 0 (si todos los artículos están bien) a 100 (si todos los artículos están defectuosos). La variable es discreta ya que cuenta ocurrencias en un conjunto finito.

3. **Llegadas de Clientes a una Tienda:**
   - **Experimento:** Observar las llegadas de clientes a una tienda durante un período de tiempo dado.
   - **Variable Aleatoria Discreta:** El número de clientes que entran a la tienda.
   - **Explicación:** Podemos sefinir $W$ como el número total de clientes que entran en la tienda en una hora. Este número puede ser cualquier entero no negativo pero no puede ser una fracción o decimal. Por lo tanto, $W$ es una variable discreta.

### Espacio Muestral

Un **espacio muestral** es el conjunto de todos los posibles resultados de un experimento aleatorio. Es un concepto fundamental porque todas las probabilidades de eventos se definen con respecto a un espacio muestral. El espacio muestral a menudo se denota con el símbolo $S$ o $\Omega$.

### Características de un Espacio Muestral:

1. **Exhaustivo:** Incluye todos los posibles resultados del experimento.
2. **Mutuamente Excluyente:** Cada resultado es distinto; no pueden ocurrir dos resultados al mismo tiempo.

### Tipos de Espacios Muestrales:

- **Espacio Muestral Discreto:** Consiste en un conjunto finito o infinito contable de resultados. Por ejemplo, los resultados de lanzar un dado.
<br><br>
- **Espacio Muestral Continuo:** Consiste en todos los valores posibles dentro de un rango o intervalo. Por ejemplo, los posibles valores de la altura de una persona.

### Ejemplo de un Espacio Muestral Discreto:

**Experimento:** Lanzar un dado de seis caras.

- **Espacio Muestral:** $S = \{1, 2, 3, 4, 5, 6\}$
<br><br>
- **Explicación:** Cada número representa un posible resultado del lanzamiento del dado. El espacio muestral es exhaustivo ya que incluye cada cara del dado, y cada resultado es mutuamente excluyente (obtener un 3, por ejemplo, excluye la posibilidad de obtener un 4 al mismo tiempo).

### Ejemplo de un Espacio Muestral Continuo:

**Experimento:** Medir la altura de adultos humanos.

- **Espacio Muestral:** Supongamos que la altura puede variar de 140 cm a 210 cm. Entonces, $S = \{x | 140 \leq x \leq 210\}$, donde $x$ representa la altura en cm.
<br><br>
- Este espacio muestral es continuo porque la altura puede tomar cualquier valor dentro del rango, no sólo mediciones discretas. Es exhaustivo para el rango especificado, y cada altura posible es un resultado distinto y mutuamente excluyente.

En ambos ejemplos, el espacio muestral proporciona la base para definir eventos y calcular probabilidades. Por ejemplo, en el lanzamiento del dado, un evento podría ser obtener un número par, que se puede definir respecto al espacio muestral.


### Definición Formal de Probabilidad

Dado un espacio muestral $S$ que representa todos los posibles resultados de un cierto experimento, y un conjunto de eventos $A, B, C, \ldots$ dentro de ese espacio muestral, la medida de probabilidad $P$ es una función que asigna a cada evento $A$ un número, $P(A)$, llamado la probabilidad de $A$, de tal manera que se satisfacen los siguientes axiomas:

1. **No negatividad:** Para cualquier evento $A$, la probabilidad de $A$ es no negativa.
   $ P(A) \geq 0 $
<br><br>
2. **Normalización:** La probabilidad del espacio muestral $S$ es 1.
   $ P(S) = 1 $
<br><br>
3. **Aditividad:** Para cualesquiera dos eventos $A$ y $B$ mutuamente excluyentes ($A$ y $B$ no pueden ocurrir al mismo tiempo), la probabilidad de la unión de $A$ y $B$ es la suma de las probabilidades de $A$ y $B$.

   $$ P(A \cup B) = P(A) + P(B) ,$$ 
   
siempre que $A$ y $B$ sean mutuamente excluyentes.

Estos axiomas se utilizan para derivar más propiedades y reglas de probabilidad, como la probabilidad del complemento de un evento: 

$$P(A^c) = 1 - P(A)$$

### Función de Masa de Probabilidad

Una **Función de Masa de Probabilidad (FMP)**  describe cómo se distribuye la probabilidad sobre los valores discretos de la variable aleatoria. En concreto, nos da la probabilidad de cada resultado.

### Puntos Clave:

1. **Definida para Variables Discretas:** Una FMP sólo tiene sentido para variables aleatorias discretas.

2. **Probabilidad de Cada Resultado:** La FMP asigna una probabilidad a cada valor posible de la variable aleatoria discreta. Si $X$ es una variable aleatoria discreta y $k$ es un valor que $X$ puede tomar, entonces la FMP $P(X = k)$ da la probabilidad de que $X$ sea igual a $k$.


3. **La Suma sobre todos los posibles valore es Igual a 1:** La suma de las probabilidades para todos los valores posibles de $X$ es igual a 1.  Matemáticamente, $\sum_{k} P(X = k) = 1$, donde la suma se realiza sobre todos los valores posibles de $X$.

### Ejemplo: lanzar un dado

Los posibles resultados son  1,2,3,4,5,6. Definamos una variable aleatoria $X$ que representa el resultado obtenido. La FMP de $X$ sería:

- $P(X = 1) = \frac{1}{6}$
- $P(X = 2) = \frac{1}{6}$
- $P(X = 3) = \frac{1}{6}$
- $P(X = 4) = \frac{1}{6}$
- $P(X = 5) = \frac{1}{6}$
- $P(X = 6) = \frac{1}{6}$

Cada resultado tiene una probabilidad igual de $\frac{1}{6}$, y la suma de estas probabilidades es 1.


La fórmula de la FMP de este experimento se conoce como **distribución uniforme discreta**, y está dada por:

$$
P(X = x_i) = \frac{1}{n}
$$

para $i = 1, 2, \ldots, n$, donde $x_i$ son los valores que $X$ puede tomar, y $n$ es el número total de valores distintos posibles de $X$ . 

La clave de esta distribución es que cada uno de los $n$ valores posibles de la variable aleatoria tiene la misma probabilidad de ocurrir.

### Ejemplo:

En nuestro caso, $n=6$, y la probabilidad de que salga cualquier cara se puede calcular como:

$$
P(X = x_i) = \frac{1}{6}
$$

para cada $x_i \in \{1, 2, 3, 4, 5, 6\}$.


Recordar: **`randint(a, b)`** es para crear una distribución uniforme discreta

```python
#Definir los límites de la distribución uniforme (1 al 7 para un dado de 6 caras)
a=1
b=7
#creamos una instancia de la distribución Uniforme Discreta: 
distribucion_dado=ss.randint(a,b)
```

### Función de Distribución

La **función de distribución acumulada (FDA)** Proporciona, para cualquier número $u$, la probabilidad de que la variable aleatoria tome un valor menor o igual a $u$. Es decir, si $X$ es una variable aleatoria y $F(x)$ es su función de distribución acumulada, entonces $$F(u) = P(X \leq u)$$.

La FDA tiene las siguientes propiedades:

1. **No decreciente:** $F(x)$ nunca disminuye a medida que $x$ aumenta. Esto significa que a medida que consideramos valores más grandes de $x$, la probabilidad acumulada no disminuye.
2. **Rango entre 0 y 1:** Para cualquier $x$, $0 \leq F(x) \leq 1$. Esto refleja el hecho de que las probabilidades siempre están entre 0 y 1.
3. **Límites en el infinito:** 
   - $$\lim_{x \to -\infty} F(x) = 0$$. Esto significa que la probabilidad de que $X$ sea menor que cualquier número extremadamente negativo es 0.
   - $$\lim_{x \to \infty} F(x) = 1$$. Esto significa que la probabilidad de que $X$ sea menor o igual que cualquier número muy grande es 1.

**Ejemplo con Variable Aleatoria Discreta**

Consideremos una variable aleatoria $X$ que representa el resultado de lanzar un dado. La FDA de $X$, $F(x)$, daría la probabilidad de obtener un resultado menor o igual a $x$. Por ejemplo, $F(3)$ sería la probabilidad de obtener un 1, 2 o 3 en un lanzamiento, la es 1/2.

**Ejemplo con Variable Aleatoria Continua**

Consideremos una variable aleatoria $X$ que representa la altura de los estudiantes en una clase, suponiendo que sigue una distribución normal. La FDA de $X$ en un valor específico, como 1.70 metros, daría la probabilidad de que un estudiante seleccionado al azar sea menor o igual a 1.70 metros de altura.

###  Bernoulli
La distribución de Bernoulli es un tipo de distribución utilizada para describir experimentos con dos posibles resultados: éxito y fracaso.

Es una distribución **discreta**, generalmente se asigna 0 (para el fracaso) y 1 (para el éxito).

Su **función de distribucion (CDF)** es:

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/f58d70d4c0e9fce4276110749998f8cea0e8c056)

Su **función de probabilidad (PMF)** es:

$$
P(X = x) = p^x(1-p)^{1-x},  \text{para x = 0,1}
$$


La distribución de **Bernoulli** la usamos cuando estamos interesados en eventos que tienen dos resultados posibles, y queremos saber la probabilidad de uno de ellos
<br><br>
Por ejemplo, si lanzamos una moneda justa, podemos modelar el resultado (àguila o sol) con una distribución de Bernoulli, donde 1 podría representar "sol" y 0 podría representar "águila".


```python
#Definimos la proba de exito con 0.7

p=0.7

dist_bernoulli=ss.bernoulli(p)
#simulo el lanzamiento de 1000 monedas
res_lanzamientos=dist_bernoulli.rvs(1000)
res_lanzamientos
```
