### pendiente:

```python
# 8.Utilizando los valores de arriba, construye una función que determine si el salario es:
# 'bajo' si salario <= percentil 25
# 'medio' si salario entre percentil 25(exluyendo) y el percentil 75 (inclusivo)
# 'alto' si salario mayor al percentil 75

#Una solución:
def clasifica_salario(x):
    if x <= emp["salario"].describe()["25%"]:
        return "bajo"
    elif x <= emp["salario"].describe()["75%"]:
        return "medio"
    else:
        return "alto"

#Desventaja: mandas a llamar más veces a la función describe()



# 9.¿Cómo aplicarías la función del punto anterior a la columna 'salario' para ver el resultado.
emp['clasificado'] = emp["salario"].apply(clasifica_salario)
```

```python
emp.head(4)
```

```python
#Otra manera de resolver 8. Al momento de hacer apply(), y = percentil 25 , z= percentil 75
y = emp["salario"].describe()['25%']
z = emp["salario"].describe()['75%']

def clasifica_salario_2(x, y, z):
    if x <= y:
        return "bajo"
    elif x <= z:
        return "medio"
    else:
        return "alto"
```

```python
#guardando info del método describe para hacer el proceso más eficiente. Implica modificar fn 'salaryClasificator'
desc = emp['salario'].describe()
emp["salario"].apply(clasifica_salario_2, args = (desc['25%'], desc['75%']))
```
