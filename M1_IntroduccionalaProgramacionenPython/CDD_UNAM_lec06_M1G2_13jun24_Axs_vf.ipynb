```python
nba = pd.read_csv('ct_nba.csv')
# insertando filas vacías (sólo para ilustrar después manejo de nulos)
nuevas_filas = pd.DataFrame(np.nan, index=[0, 1, 2], columns=nba.columns)

# concatenando y reseteando el índice
nba = pd.concat([nba, nuevas_filas], ignore_index=True)

nba.tail()
```

```python
# SI queremos quitar los renglones que tienen Nan en TODAS las columnas y quiero hacerlo permanentemente
nba.dropna(how='all', inplace=True)
nba.tail()
```

```python
nba.info()
```

```python
#Eliminar COLUMNAS con al menos un valor nulo:
nba.dropna(axis=1) #me quita Position y College
```

```python
#equivalente a lo anterior:
nba.dropna(axis='columns')
```

```python
# Quiero el NÚMERO de registros NO nulos en la variable College
nba['College'].notna().sum()
```
