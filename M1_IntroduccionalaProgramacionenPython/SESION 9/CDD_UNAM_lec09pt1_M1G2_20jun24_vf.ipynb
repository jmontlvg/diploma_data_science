## `For` loops
Usamos un `for` cuando conocemos a priori el número de veces que deseamos correr un bloque de código. 

La sintaxis para un *for loop* en python es la siguiente:

```python
for elemento in secuencia:
    # hacer algo con elemento
```

* `secuencia` es un iterable: todo aquello que contenga un índice (lista, string, diccionario, etc.)
* `elemento`: es una variable que toma como valor cada elemento dentro de `secuencia`.
* Todo bloque de código con  sangría después del `for`se repite por cada `elemento`dentro de `secuencia`.

* Cuándo Usar: Ideal para iterar cuando se conoce el número de iteraciones o para iterar directamente sobre los elementos de una secuencia.
* Ejemplos:
    * Iterar sobre una lista:

````python
frutas = ["manzana", "banana", "cereza"]
for fruta in frutas:
    print(fruta)
````

````python
palabra = 'palabra'

for letra in palabra:
    print(letra)
````

````python
for i in range(6):
    print (i)
````

````python
for i in range(6):
    for k in range(5):
        print(i,k)
````

## `While` loops

* Qué Es: Un bucle while en Python ejecuta repetidamente la instrucción objetivo mientras la condición dada sea verdadera.
* Sintaxis:
```python
while condicion:
    # hacer algo
```

* Cuándo Usar: Mejor cuando no se conoce el número de iteraciones de antemano y desea continuar bucleando mientras se cumpla una condición.
* Ejemplo:
    * Una cuenta regresiva simple:



````python
cuenta = 5
while cuenta > 0:
    print(cuenta)
    cuenta-=1        #cuenta = cuenta -1 es equivalente a cuenta-=1
````

## `Do While` loops  (no hay en Python : ) )

* Se puede emular su funcionalidad usando un ciclo de tipo `while`
* Cómo Emular:
    * La idea es ejecutar el cuerpo del ciclo una vez y luego continuar con base en una condición.
* Sintaxis:
```python
while True:
    # hacer algo
    if not condicion:
        break


* Ejemplo:
    * Emular 'do while' para solicitar al usuario hasta condición de finalizar:

````python
libros = []
print("ingresa nombres de libros que te gustan. Escribe 'fin' para terminar el proceso")
while True:  #por definición la condición en verdadera.
    libro = input("libro: ")
    if libro != 'fin':
        libros.append(libro)
    else:
        print('fin')
        break
````

### Recordando range()

**Rangos**
La forma de crear un rango de números dentro de python es mediante la función `range`, la cual puede ser usada de 3
maneras:
* `range(a)` crea un rango de valores de `0`hasta `a-1`
* `range(a,b)` crea un rango de valores de `a` hasta `b-1`
* `range(a,b, s)` crea un rango de valores de `a` hasta `b-1` dando saltos `s`
